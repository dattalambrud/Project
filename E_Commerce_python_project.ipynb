```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\Datta\Downloads\E-commerce.csv",encoding='latin 1')
```

```python
data['City'].value_counts()[:5]
```

```python
#changing datatype object to datatime
data['Order Date']=pd.to_datetime(data['Order Date'])
```

```python
#separating years
data['order year']=data['Order Date'].dt.year
```

```python
#order counts
data['order year'].value_counts()
```

```python
#showing every yaers sales
df=sns.countplot(x='order year', data=data,color='orange')
plt.title('showing years of selling')
for bars in df.containers:
    df.bar_label(bars)
plt.show()
```

```python
#find out buyer segment
data['Segment'].value_counts()
```

```python
#create countplot for buyer segment
df=sns.countplot(data=data,x='Segment',color='red')
plt.title('count of buyer segment')
plt.figure(figsize=(10,5))
for bars in df.containers:
    df.bar_label(bars)
```

```python
data['Ship Mode'].value_counts()
```

```python
# create pie chart for ship mode to figure out consumer ship day
df=data['Ship Mode'].value_counts()
plt.pie(df,labels=df.index,autopct='%0.2f%%')
plt.legend(loc=1)
plt.show()
```

```python
#distribution of orders by category and ship mode
plt.figure(figsize=(6,5))
df=sns.countplot(data=data,x='Category',hue='Ship Mode')
for bars in df.containers:
    df.bar_label(bars)
```

```python
#correlation between sales and profit
plt.scatter(data['Category'],data['Sub-Category'])
plt.title('scatter plot')
plt.show()
```

```python
#showing profit category wise
df=sns.barplot(x='Category',y='Profit',data=data,estimator='sum',color='purple')
plt.show()
```
